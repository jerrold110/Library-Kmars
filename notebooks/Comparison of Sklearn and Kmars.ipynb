## Create blobs

```python
import numpy as np
from sklearn.datasets import make_blobs

n_samples = 1500
random_state = 170
transformation = [[0.60834549, -0.63667341], [-0.40887718, 0.85253229]]

X, y = make_blobs(n_samples=n_samples, random_state=random_state)
# Anisotropic blobs
X_aniso = np.dot(X, transformation)  
# Unequal variance
X_varied, y_varied = make_blobs(n_samples=n_samples, cluster_std=[1.0, 2.5, 0.5], random_state=random_state)
# Unevenly sized blobs
X_filtered = np.vstack((X[y == 0][:500], X[y == 1][:100], X[y == 2][:10]))  
y_filtered = [0] * 500 + [1] * 100 + [2] * 10
```

```python
import matplotlib.pyplot as plt
```

## Scikit-learn KMeans (Euclidean)

```python
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

km1 = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=0.0001, random_state=0).fit(X)
km2 = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=0.0001, random_state=0).fit(X_aniso)
km3 = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=0.0001, random_state=0).fit(X_varied)
km4 = KMeans(n_clusters=3, init='k-means++', n_init=10, max_iter=300, tol=0.0001, random_state=0).fit(X_filtered)

fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(8, 8))

axs[0, 0].scatter(X[:, 0], X[:, 1], c=km1.labels_)
axs[0, 0].scatter(km1.cluster_centers_[:,0], km1.cluster_centers_[:,1], marker="x", color='r')
axs[0, 0].set_title("Mixture of Gaussian Blobs")

axs[0, 1].scatter(X_aniso[:, 0], X_aniso[:, 1], c=km2.labels_)
axs[0, 1].scatter(km2.cluster_centers_[:,0], km2.cluster_centers_[:,1], marker="x", color='r')
axs[0, 1].set_title("Anisotropically Distributed Blobs")

axs[1, 0].scatter(X_varied[:, 0], X_varied[:, 1], c=km3.labels_)
axs[1, 0].scatter(km3.cluster_centers_[:,0], km3.cluster_centers_[:,1], marker="x", color='r')
axs[1, 0].set_title("Unequal Variance")

axs[1, 1].scatter(X_filtered[:, 0], X_filtered[:, 1], c=km4.labels_)
axs[1, 1].scatter(km4.cluster_centers_[:,0], km4.cluster_centers_[:,1], marker="x", color='r')
axs[1, 1].set_title("Unevenly Sized Blobs")

plt.show()
```
